```python
import numpy as np
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
p = 120
```

```python
time = np.arange(p)
```

```python
time
```

```python
series = np.sin(2 * np.pi * time / 12) + 0.5 * time + np.random.normal(p)
```

```python
series
```

```python
decomposition = seasonal_decompose(series, period = 12)
```

```python
trend = decomposition.trend
```

```python
trend
```

```python
season = decomposition.seasonal
```

```python
season
```

```python
residual = decomposition.resid
```

```python
residual
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))
plt.subplot(411)
plt.plot(time, series, label = 'Data')
plt.legend(loc = 'best')
plt.subplot(412)
plt.plot(time, trend, label = 'Trend')
plt.legend(loc = 'best')
plt.subplot(413)
plt.plot(time, residual, label = 'Residual')
plt.legend(loc = 'best')
plt.subplot(414)
plt.plot(time, season, label = 'Season')
plt.legend(loc = 'best')
```
