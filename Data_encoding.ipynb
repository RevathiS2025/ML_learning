```python
df1 = df.dropna()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

```

```python
df1['Gender'] = le.fit_transform(df1['Gender'])
```

```python
df1
```

```python
df1['Married'] = le.fit_transform(df1['Married'])
df1
```

```python
df2 = pd.get_dummies(df1,dtype=int)
```

```python
df2
```

```python
df3 = df1.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder

oc = OneHotEncoder()
```

```python
df3_encoded = oc.fit_transform(df3[['Gender']]).toarray()
```

```python
df3_encoded
```

```python
encoded_Gender = pd.DataFrame(df3_encoded)
```

```python
df3=pd.concat([df3,encoded_Gender],axis=1)
```
