```python
df2 = df.copy()
```

```python
df2=df2.drop('Loan_ID', axis = 1)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df2['Gender'] = le.fit_transform(df2['Gender'])
```

```python
df2['Married'] = le.fit_transform(df2['Married'])
```

```python
df2['Dependents'] = le.fit_transform(df2['Dependents'])
```

```python
df2['Education'] = le.fit_transform(df2['Education'])
```

```python
df2['Self_Employed'] = le.fit_transform(df2['Self_Employed'])
```

```python
df2['Property_Area'] = le.fit_transform(df2['Property_Area'])
```

```python
df2['Loan_Status'] = le.fit_transform(df2['Loan_Status'])
```

```python
df2
```

```python
df2.info()
```

```python
x = df2.iloc[:,:-1]
y = df2.iloc[:,11:12]
```

```python
x
```

```python
from sklearn import preprocessing
norm_x = preprocessing.normalize(x)
```

```python
norm_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(norm_x,y,test_size=0.3, random_state=3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```
