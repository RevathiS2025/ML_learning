```python
import pandas as pd
import numpy as np

```

```python
from sklearn.datasets import fetch_california_housing
```

```python
californiaHousing = fetch_california_housing()
```

```python
californiaHousing
```

```python

df = pd.DataFrame(californiaHousing.data, columns= californiaHousing.feature_names)
```

```python
df
```

```python
df['Price'] = californiaHousing.target
```

```python
x = df[['MedInc','HouseAge','AveRooms']]
```

```python
x
```

```python
y = df[['Price']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=179)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.linear_model import BayesianRidge #Naive Bayes Regression
```

```python
lr = LinearRegression()
svr = SVR(kernel='linear')
br = BayesianRidge()
```

```python
lr.fit(x_train, y_train)
```

```python
svr.fit(x_train, y_train)
```

```python
br.fit(x_train, y_train)
```

```python
y_predict1 = lr.predict(x_test)
```

```python
y_predict2 = svr.predict(x_test)
```

```python
y_predict3 = br.predict(x_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=5)
```

```python
x_poly = poly.fit_transform(x_train)
```

```python
x_poly
```

```python
x_train
```

```python
lr.fit(x_poly,y_train)
```

```python
result2 = poly.transform(x_test)
```

```python
y_predict4 = lr.predict(result2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

Result analysis of **MLR, SVR, BR** respectively

```python
mae_mlr = mean_absolute_error(y_test,y_predict4) #MLR
mse_mlr = mean_squared_error(y_test,y_predict4)
r2_mlr = r2_score(y_test,y_predict4)
```
