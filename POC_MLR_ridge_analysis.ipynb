```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
californiaHousing = fetch_california_housing()
```

```python
californiaHousing
```

```python
df = pd.DataFrame(californiaHousing.data, columns= californiaHousing.feature_names)
```

```python
df
```

```python
df['Price'] = californiaHousing.target
```

```python
df.head()
```

```python
df.tail()
```

```python
df.iloc[10010:10050]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x = df.drop(columns=['Price', 'Latitude','Longitude'])
```

```python
x
```

```python
y = df[['Price']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=179)
```

```python
mlr = LinearRegression()
rd= Ridge(alpha=0.001)
```

```python
mlr.fit(x_train, y_train)
```

```python
rd.fit(x_train, y_train)
```

```python
y_predict1 = mlr.predict(x_test)
```

```python
y_predict2 = rd.predict(x_test)
```

```python
x_train
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

Result analysis of **MLR, Ridge** respectively

```python
mae_mlr = mean_absolute_error(y_test,y_predict1) #MLR
mse_mlr = mean_squared_error(y_test,y_predict1)
r2_mlr = r2_score(y_test,y_predict1)
```

```python
mae_mlr
```

```python
mse_mlr
```

```python
r2_mlr
```

```python
mae_rd = mean_absolute_error(y_test,y_predict2) #SVR
mse_rd = mean_squared_error(y_test,y_predict2)
r2_rd = r2_score(y_test,y_predict2)
```

```python
mae_rd
```

```python
mse_rd
```

```python
r2_rd
```

```python
# Ensure 1D arrays
y_test = np.ravel(y_test)
y_predict1 = np.ravel(y_predict1)
y_predict2 = np.ravel(y_predict2)
```

```python

# Scatter plot of actual vs predicted values
plt.scatter(y_test, y_predict1, color='blue', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)],
         color='red', linestyle='--', linewidth=2)
plt.title('Multi Linear Regression')
plt.xlabel('Actual')
plt.ylabel('Predict')
plt.show()
```

```python

# Scatter plot of actual vs predicted values
plt.scatter(y_test, y_predict2, color='blue', alpha=0.6)

# Perfect prediction reference line (diagonal)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)],
         color='red', linestyle='--', linewidth=2)
plt.title('Ridge Regression')
plt.xlabel('Actual')
plt.ylabel('Predict')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
mlr.fit(X_train_scaled, y_train)
y_predict3 = mlr.predict(X_test_scaled)
```

```python
mae_scaled = mean_absolute_error(y_test,y_predict3)
mse_scaled = mean_squared_error(y_test,y_predict3)
r2_scaled = r2_score(y_test,y_predict3)
```
