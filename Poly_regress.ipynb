```python
# x, y split
x=df.iloc[:,:-1]
y=df.iloc[:,1]
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x, lin_reg.predict(x), color='red')
plt.title('Gaming Data (Linear Regression)')
plt.xlabel('Gaming Steps')
plt.ylabel('Points')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=9) # Tranforming
X_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x, lin_reg2.predict(poly_reg.fit_transform(x)), color='red')
plt.title('Gaming Data (Polynomial Regression)')
plt.xlabel('Gaming Steps')
plt.ylabel('Points')
plt.show()
```
